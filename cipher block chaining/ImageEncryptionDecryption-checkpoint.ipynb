```python
from Crypto.Cipher import AES
from PIL import Image
import base64
import os
import sys


IV_LEN = 16
BLOCK_LEN = 16
key = "aaaabbbbccccdddd"
key = str.encode(key)
def usage():
    """
    ./aes.py encrypt keyfile image.png
    ./aes.py decrypt keyfile enc_image.png
    """
    if len(sys.argv) != 4:
        print("Usage: \n" \
              + "./aes.py encrypt keyfile image.png \n" \
              + "./aes.py decrypt keyfile enc_image.png")
    else:
        if sys.argv[1] == 'encrypt':
            encrypt(sys.argv[2], sys.argv[3])
        elif sys.argv[1] == 'decrypt':
            decrypt(sys.argv[2], sys.argv[3])
        else:
            print("Invalid operation!")

def encrypt(key, imagefile):
    """
    Encrypts image with given key
    """
    
    if len(key) % 16 != 0:
        print('Key size must be a multiple of 16')
        sys.exit(1)

    iv = os.urandom(IV_LEN)
    with open(imagefile, 'rb') as image_file:
        image_str = base64.b64encode(image_file.read())
    #print("1111111111")
    #print(image_str)
    padding = BLOCK_LEN - len(image_str) % BLOCK_LEN
    image_str += bytes([padding]) * padding
    print(len(image_str))
    #print("------------------")
    aes = AES.new(key, AES.MODE_CBC, iv)
    enc = aes.encrypt(image_str)
    #print(image_str)
    
    with open('iv', 'wb') as iv_file:
        iv_file.write(iv)
    
    with open('enc_'+imagefile, 'wb') as image_file:
        image_file.write(enc)

    size = (len(enc)/3) ** (1 / 2)
    encrypted_img = Image.frombuffer("RGB", (int(size), int(size)), enc, 'raw', "RGB", 0, 1)
    encrypted_img.save('encrypted.jpg', 'JPEG')
    return image_str


def decrypt(key, imagefile):
    """
    Decrypts image with given key
    """

    if len(key) % 16 != 0:
        print('Key size must be a multiple of 16')
        sys.exit(1)
    
    with open(imagefile, 'rb') as image_file:
        enc_str = image_file.read()

    with open('iv', 'rb') as iv_file:
        iv = iv_file.read()
    
    aes = AES.new(key, AES.MODE_CBC, iv)
    data = aes.decrypt(enc_str)
    data = data[:-data[-1]]
    
    with open('decrypted_'+imagefile[4:], 'wb') as image_file:
        image_file.write(base64.b64decode(data))
        
    

img = "input.jpg"
x = encrypt(key, img)
decrypt(key,'enc_'+img)
```
